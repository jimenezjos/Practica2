```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fx = lambda x: 2*(x**3) + (x**2) - 13*x + 6
dx = lambda x: 6*(x**2) + 2*x - 13
xi=2.5
columnas=["interaciones","xi","f(xi)","f´(xi)","xi+1","error"]
datos=[[0,0,0,0,xi,0]]
df = pd.DataFrame(datos,columns=columnas)
i=1
tolerancia = 0.00001
error = 1
while error>tolerancia:

  Raiz = xi-(fx(xi)/dx(xi))
  error = np.abs((Raiz-xi)/Raiz)
  df = df.append({
      "interacion":i,
      "xi":xi,
      "f(xi)":fx(xi),
      "f´(xi)":dx(xi),
      "xi+1":Raiz,
      "error":error
  },ignore_index=True)
  xi = Raiz
  i+=1
print("La raiz es: ",Raiz)  
print("El valor de f(x): ",fx(Raiz))
df.head(15)
```
